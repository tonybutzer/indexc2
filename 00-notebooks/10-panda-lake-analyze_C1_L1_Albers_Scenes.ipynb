```python
import pandas as pd
import glob

path = '../prepare/analysis' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    print(filename)
    #df = pd.read_csv(filename, index_col=None, header=0)
    df = pd.read_csv(filename, index_col=None, header=None)
    li.append(df)
    

pf = pd.concat(li, axis=0, ignore_index=True)
```

```python
pf.head()
```

```python
pf.columns = ['Collection', 'Level', 'Projection', 'Sensor', 'Year', 'Path', 'Row', 'Directory', 'XML_File']
```

```python
pf.head()
```

```python
pf.describe()
```

```python
#nf = pf.query('Path'=='10' and 'Row'=='28')

nf = pf.loc[pf['Path']==46]

hayden_pf = nf.loc[nf['Row']==28]

hayden_pf
```

```python
hayden_pf.describe()
```

```python
rows_year = hayden_pf.groupby('Year').count()
```

```python
year_count = rows_year[['Directory', 'XML_File']]
```
